```python
# Data
departments = ['HR', 'IT', 'Sales']
satisfaction_levels = ['Satisfied', 'Neutral', 'Dissatisfied']


# Observed Frequencies (O)
observed = {
    'HR': {'Satisfied': 20, 'Neutral': 5, 'Dissatisfied': 5},
    'IT': {'Satisfied': 25, 'Neutral': 10, 'Dissatisfied': 5},
    'Sales': {'Satisfied': 15, 'Neutral': 5, 'Dissatisfied': 10}
}

# Row Totals and Column Totals
row_totals = {}
column_totals = {'Satisfied': 0, 'Neutral': 0, 'Dissatisfied': 0}
grand_total = 0

for dept in departments:
    total = sum(observed[dept].values())
    row_totals[dept] = total
    grand_total += total
    for level in satisfaction_levels:
        column_totals[level] += observed[dept][level]

# Expected Frequencies (E)
expected = {}
for dept in departments:
    expected[dept] = {}
    for level in satisfaction_levels:
        expected_freq = (row_totals[dept] * column_totals[level]) / grand_total
        expected[dept][level] = expected_freq

# Chi-Square Calculation
chi_square = 0
for dept in departments:
    for level in satisfaction_levels:
        O = observed[dept][level]
        E = expected[dept][level]
        chi_square += (O - E) ** 2 / E

# Degrees of Freedom
df = (len(departments) - 1) * (len(satisfaction_levels) - 1)

# Critical Chi-Square Value
chi_square_critical = 9.488

# Output the results
print("Observed Frequencies:")
for dept in departments:
    print(f"{dept}: {observed[dept]}")

print("\nExpected Frequencies:")
for dept in departments:
    print(f"{dept}: {expected[dept]}")

print(f"\nChi-Square Statistic: {chi_square:.4f}")
print(f"Degrees of Freedom: {df}")
print(f"Critical Chi-Square Value: {chi_square_critical}")

# Decision
if chi_square > chi_square_critical:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
# Data
Gender = ['Male', 'Female']
Product = ['Yes', 'No']

# Observed Frequencies (O)
observed = {
    'Male': {'Yes': 30, 'No': 20},
    'Female': {'Yes': 40, 'No': 10}
}

# Row Totals, Column Totals, and Grand Total
row_totals = {}
column_totals = {'Yes': 0, 'No': 0}
grand_total = 0

for i in Gender:
    total = sum(observed[i].values())
    row_totals[i] = total
    grand_total += total
    for j in Product:
        column_totals[j] += observed[i][j]

# Expected Frequencies (E)
expected = {}
for i in Gender:
    expected[i] = {}
    for j in Product:
        expected_freq = (row_totals[i] * column_totals[j]) / grand_total
        expected[i][j] = expected_freq

# Chi-Square Calculation with Individual Cell Contributions
chi_square = 0
chi_square_contributions = {}
for i in Gender:
    chi_square_contributions[i] = {}
    for j in Product:
        O = observed[i][j]
        E = expected[i][j]
        chi_square_contrib = (O - E) ** 2 / E
        chi_square += chi_square_contrib
        chi_square_contributions[i][j] = chi_square_contrib

# Degrees of Freedom
df = (len(Gender) - 1) * (len(Product) - 1)

# Critical Chi-Square Value
chi_square_critical = 3.841 

# Output Observed and Expected Frequencies, and Chi-Square Contributions
print("Observed Frequencies:")
for i in Gender:
    print(f"{i}: {observed[i]}")

print("\nExpected Frequencies:")
for i in Gender:
    print(f"{i}: {expected[i]}")

print("\nChi-Square Contributions for Each Cell:")
for i in Gender:
    for j in Product:
        print(f"Contribution for {i}-{j}: {chi_square_contributions[i][j]:.4f}")

# Output Chi-Square Statistic and Decision
print(f"\nChi-Square Statistic: {chi_square:.4f}")
print(f"Degrees of Freedom: {df}")
print(f"Critical Chi-Square Value: {chi_square_critical}")

# Decision
if chi_square > chi_square_critical:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```
